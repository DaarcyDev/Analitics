```python
# pruebas de bondad y ajuste
# --saphiro-wilk
# --kolmogorov-smirnov
from scipy.stats import shapiro
from scipy.stats import kstest,norm

sample = df["Base Salary"].sample(n=50)

print("shapiro")
stats,p_value = shapiro(sample)

print("stats",stats)
print("p_value",p_value)
IC=0.05
if(p_value>IC):
    print("Acepta la H0: la distriucion es normal")
else:
    print("Rechaza la H1: la distriucion no es normal")

print("")

print("kstest")
stats,p_value = kstest(sample, norm.cdf)
print("stats",stats)
print("p_value",p_value)
IC=0.05
if(p_value>IC):
    print("Acepta la H0: la distriucion es normal")
else:
    print("Rechaza la H1: la distriucion no es normal")

print("")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings("ignore")
cwd = os.getcwd()
os.chdir("/home/daarcydev/Desktop/Proyects/Analitica/")

population = pd.read_csv("employee_salary.csv",header=0,sep=",")

populatio_base_salary =  population["Base Salary"]
sns.displot(populatio_base_salary)
plt.xlabel('populatio base salary')
plt.ylabel('Probability density')

mean=[]

def calc_sample_mean(data,sample_size,no_sample_means):
    for i in range(no_sample_means):
        sample=data.sample(n=sample_size)
        sample_mean = sample.mean()
        mean.append(sample_mean)
    return (mean)

mean_2 = calc_sample_mean(data = populatio_base_salary, sample_size=2, no_sample_means=500)

sns.displot(mean_2, color='b')
plt.xlabel('sample base salary(sample size = 2)')
plt.ylabel('Probability density')


mean_3 = calc_sample_mean(data = populatio_base_salary, sample_size=5, no_sample_means=500)

sns.displot(mean_3, color='r')
plt.xlabel('sample base salary(sample size = 5)')
plt.ylabel('Probability density')

mean_10 = calc_sample_mean(data = populatio_base_salary, sample_size=10, no_sample_means=500)

sns.displot(mean_10, color='y')
plt.xlabel('sample base salary(sample size = 10)')
plt.ylabel('Probability density')

mean_20 = calc_sample_mean(data = populatio_base_salary, sample_size=20, no_sample_means=500)

sns.displot(mean_20, color='green')
plt.xlabel('sample base salary(sample size = 20)')
plt.ylabel('Probability density')

mean_30 = calc_sample_mean(data = populatio_base_salary, sample_size=30, no_sample_means=500)

sns.displot(mean_30, color='brown')
plt.xlabel('sample base salary(sample size = 30)')
plt.ylabel('Probability density')

mean_50 = calc_sample_mean(data = populatio_base_salary, sample_size=50, no_sample_means=500)

sns.displot(mean_50, color='magenta')
plt.xlabel('sample base salary(sample size = 50)')
plt.ylabel('Probability density')
```

```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, accuracy_score, f1_score, confusion_matrix, classification_report
import seaborn as sns

df = pd.read_csv('iris.csv.txt', header = 0,sep =',')

df.head(5)

features = ["sepal_length","sepal_width","petal_length","petal_width"]
label = "species"

sle = LabelEncoder()
df['sle'] = sle.fit_transform(df[label])

xtrain, xtest, ytrain, ytest = train_test_split(df[features],df.sle,test_size=0.2, random_state=0)

print(xtrain.shape," ->0.80")
print(xtest.shape," ->0.20")

print(len(ytrain))
print(len(ytest))

svm= SVC()
modelo = svm.fit(xtrain,ytrain)
prediccion= modelo.predict(xtest)
```

```python
svm = SVC( probability=True)
modelo = svm.fit(xtrain,ytrain)
prediccion = modelo.predict(xtest)
```

```python
print(df['species'].value_counts())
print('score ',modelo.score(xtest, ytest))

print("recall_score ", recall_score(prediccion, ytest, average='micro'))
print("accuracy_score ", accuracy_score(prediccion, ytest))
print("f1_score ", f1_score(prediccion, ytest, average='micro'))
print("confusion_matrix \n", confusion_matrix(prediccion, ytest))
print("classification_report \n", classification_report(prediccion, ytest))

cf = confusion_matrix(prediccion, ytest)
sns.heatmap(cf, annot=True)
```
