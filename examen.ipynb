```python
import pandas as pd
import os
import missingno as msno
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet

df = pd.read_csv('boston_train.csv', header = 0,sep =',')

df
```

```python

nulos_por_columna = df.isnull().sum()

print("Valores nulos por columna:")
print(nulos_por_columna)
```

```python


# df_sin_nulos = df.fillna(0)
df_sin_nulos = df.dropna(axis=0)
df = df_sin_nulos  

df = df.replace('?', df.mode().iloc[0])
(df == '?').any()


```

```python
features = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
label = 'MEDV'
le=LabelEncoder()
y=le.fit_transform(df[label])
xtrain,xtest,ytrain,ytest=train_test_split(df[features],y,test_size=0.20,random_state=0)
print('xtrain.shape:')
print(xtrain.shape)

print('\nlen(ytrain):')
print(len(ytrain))

print('\nxtest.shape:')
print(xtest.shape)

print('\nlen(ytest):')
print(len(ytest))
```

```python
svm = LinearRegression()
modelo = svm.fit(xtrain, ytrain)
prediccion = modelo.predict(xtest)
prediccion_list = prediccion.tolist()
ytest_list = ytest.tolist()

i = 0
for pred in prediccion_list:
	# print(round(pred), '->', ytest[i])
	i += 1
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print('score', modelo.score(xtest, ytest))
print('mean_squared_error', mean_squared_error(ytest, prediccion))
print('mean_absolute_error', mean_absolute_error(ytest, prediccion))
```

```python
from sklearn.linear_model import Ridge
svm = Ridge(max_iter=1000)
modelo = svm.fit(xtrain, ytrain)
prediccion = modelo.predict(xtest)
prediccion_list = prediccion.tolist()
ytest_list = ytest.tolist()

i = 0
for pred in prediccion_list:
	# print(round(pred), '->', ytest[i])
	i += 1
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print('score', modelo.score(xtest, ytest))
print('mean_squared_error', mean_squared_error(ytest, prediccion))
print('mean_absolute_error', mean_absolute_error(ytest, prediccion))
```
